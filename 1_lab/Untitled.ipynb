```python
import numpy as np
import matplotlib.pyplot as plt

# Define range around x0 = 0
x0 = 0
x = np.linspace(-1, 1, 400)
h = x - x0

# Compute terms (x - x0)^m and (x - x0)^(m+1)
m = 2
term_m = h ** m
term_m1 = h ** (m + 1)

# Plot
plt.figure(figsize=(8, 6))
plt.plot(x, term_m, label=r'$(x - x_0)^m$', linewidth=2)
plt.plot(x, term_m1, label=r'$(x - x_0)^{m+1}$', linestyle='--', linewidth=2)
plt.axvline(0, color='gray', linestyle=':', linewidth=1)
plt.title(r'Comparison of $(x - x_0)^m$ and $(x - x_0)^{m+1}$ near $x_0=0$', fontsize=14)
plt.xlabel(r'$x$', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```
